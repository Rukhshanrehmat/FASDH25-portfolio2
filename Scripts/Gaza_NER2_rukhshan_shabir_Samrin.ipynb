## Creating the pipeline

Download the English language model and build the pipeline (we specify that it should only tokenize the text, separate multiword tokens and perform Named Entity Recognition):


```python
# Download the English Model
stanza.download("en")
# write a code to load the English pipeline for NER
nlp = stanza.Pipeline("en", processors="tokenize,ner")
```

We can now loop through the articles in the folder as we did when we were using regex to find filenames:

```python
import os
import re

# write a code to set path to which folder to use
articles_folder = "/content/FASDH25-portfolio2/articles"
# write a code to list all files from the folder
files = os.listdir(articles_folder)
# write a code to only select articles from January 2024
# check CHATGPT Solution 2B.1
january_files = []
for file in files:     # loop through all files in  folder
  if file.startswith("2024-01"):
    january_files.append(file) # this code will add files to january file list.
# print the selected articles
print(f"number of articles selected for january: {len(january_files)}")

# write a code to create an empty dictionary to store place names and their counts
place_counts = {}
# write a code to loop through each file that was selected from January 2024
for filename in january_files:
    path = os.path.join(articles_folder, filename)
    # write a code to open the file and read its content
    with open(path, encoding="utf-8") as file:
        text = file.read()
        doc = nlp(text)
        # write a code to loop through each entities detected in the text
        for entity in doc.entities:
            if entity.type in ["GPE", "LOC"]:  # Only consider places
                place = entity.text
                # write a code to count the place name (add 1 if already exists, or start at 1)
        # this code is adapted from slides
                if place in place_counts:
                    place_counts[place] += 1
                else:
                    place_counts[place] = 1
#following chunks of code are adapted by lecture slides and with the help of chat gpt: check CHATGPT solution 2B.2
# Print the final dictionary of place names and their counts
print(place_counts)

# Create a new dictionary to store cleaned place names
cleaned_place_counts = {}

for place, count in place_counts.items():
    # write a code to lowercase everything for consistency
    place = place.lower()

    # write a code to remove possessives like 's or ’s
    place = re.sub(r"[’'`]s\b", "", place)

    # write a code to remove all punctuation
    place = re.sub(r"[^\w\s]", "", place)

    # write a code to remove leading 'the' if it appears
    place = re.sub(r"^the\s+", "", place)

    # write a code to capitalize each word (title case)
    place = place.title()

    # Step 6: Merge counts
    if place in cleaned_place_counts:
        cleaned_place_counts[place] += count
    else:
        cleaned_place_counts[place] = count

# Print cleaned and normalized results
for place, count in cleaned_place_counts.items():
    print(f"{place}: {count}")
```

### Storing data in a tsv file

We can now store the counts in a tsv file, so we can reuse it in a different script.

Let's create a tsv file with two columns: "name" and "frequency".
We'll create the tsv file in two steps:

1. we create the header: that is, the column names, separated by tabs
2. we loop through all the place names, and we create a new row in the table for each place. Each row will contain the place name and its frequency, separated by a tab. Each row will have to start on a new line, so we'll also have to add a newline character \n to the row; should we add it at the beginning or end of the line, or both?

Fill in the blanks:

```python
filename = "ner_counts.tsv"

# Open the file in writing mode and with UTF-8 encoding
with open(filename, mode="w", encoding="utf-8") as file:
    # Write header row
    header = "Place\tCount\n"
    file.write(header)

    # Loop through the cleaned place counts and write each row
    for place, count in cleaned_place_counts.items():
        row = f"{place}\t{count}\n"
        file.write(row)
```
